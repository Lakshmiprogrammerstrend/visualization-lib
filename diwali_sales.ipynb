```python
#library
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
#read csv
df = pd.read_csv('/content/Diwali Sales Data.csv',encoding="unicode_escape")
```

```python
print(df.shape)
```

```python
print(df.head(10))
```

```python
print(df.info())
```

```python
df.drop(['Status','unnamed1'],axis=1,inplace=True)
```

```python
df.info()
```

```python
pd.isnull(df).sum()
```

```python
df.dropna(inplace=True)
```

```python
pd.isnull(df).sum()
```

```python
df['Amount'] = df['Amount'].astype(int)
```

```python
df['Amount'].dtypes
```

```python
df.columns
```

```python
df.rename(columns={"Marital_Status":"M_S"})
```

```python
df.describe()
```

```python
df[['Age','Orders','Amount']].describe()
```

```python
ax =sns.countplot(x='Gender',data = df,palette = "rocket")
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sale_gen = df.groupby(["Gender"],as_index = False)["Amount"].sum().sort_values(by="Amount",ascending=False)
sns_plot = sns.barplot(x = "Gender",y = "Amount",data = sale_gen)
for bars in sns_plot.containers:
    sns_plot.bar_label(bars)
```

```python
df1 = sns.countplot(x="Age Group",hue ="Gender",data = df)
for bars in df1.containers:
  df1.bar_label(bars)
```

```python
sale_age = df.groupby(['Age Group'],as_index = False)['Amount'].sum().sort_values(by ="Amount",ascending = False)
sns.set(rc = {'figure.figsize':(10,5)})
k = sns.barplot(x="Age Group",y='Amount',data = sale_age)
for bars in k.containers:
  k.bar_label(bars)
```

```python
sale_sate = df.groupby(['State'],as_index = False)['Orders'].sum().sort_values(by='Orders',ascending = False).head(10)
sns.set(rc={"figure.figsize":(18,5)})
sal_2 = sns.barplot(x ="State",y= 'Orders',data = sale_sate)
for bars in sal_2.containers:
  sal_2.bar_label(bars)
```

```python
sale_state = df.groupby(['State'],as_index = False)['Amount'].sum().sort_values(by = 'Amount',ascending = False).head(10)
sns.set(rc={"figure.figsize":(18,5)})
ad = sns.barplot(x = 'State',y= 'Amount',data = sale_state, palette = "rocket")
for labels in ad.containers:
  ad.bar_label(labels)
```

```python
ax = sns.countplot(x="Marital_Status",data = df)
sns.set(rc={"figure.figsize":(5,2)})
for bars in ax.containers:
  ax.bar_label(bars)
```

```python
sale_satus=df.groupby(["Marital_Status","Gender"],as_index = False)['Amount'].sum().sort_values(by ='Amount',ascending = False )
sns.barplot(data = sale_satus,x='Marital_Status',y = 'Amount',hue = 'Gender')
```

```python
sns.set(rc={"figure.figsize":(18,5)})
ax = sns.countplot(x = 'Occupation',data = df )

for bars in ax.containers:
  ax.bar_label(bars)
```

```python
slae_occ= df.groupby(['Occupation'],as_index = False)['Amount'].sum().sort_values(by="Amount",ascending= False)
sns.set(rc={"figure.figsize":(20,5)})
dp = sns.barplot(x='Occupation',y = 'Amount',data =slae_occ )
for bars in dp.containers:
  dp.bar_label(bars)
```

```python
sns.set(rc={"figure.figsize":(25,5)})
kick = sns.countplot(data = df,x ='Product_Category')
for bars in kick.containers:
  kick.bar_label(bars)
```

```python
sns.set(rc={"figure.figsize":(25,5)})
sale_state = df.groupby(['Product_Category'],as_index = False)['Amount'].sum().sort_values(by="Amount",ascending= False).head(10)
dk = sns.barplot(data = sale_state ,x = 'Product_Category',y = 'Amount')
for bars in dk.containers:
  dk.bar_label(bars)
```

```python
sales_state = df.groupby(['Product_ID'],as_index=False)["Orders"].sum().sort_values(by = 'Orders',ascending = False).head(10)
dop = sns.barplot(data =sales_state,x = 'Product_ID',y= 'Orders' )
for bars in dop.containers:
  dp.bar_label(bars)
```
